```python
import os
import xml.etree.ElementTree as ET
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
!pwd
```

```python
path="./xml"
```

```python
all_path=os.listdir(path)
len(all_path)
```

```python
i=0
for ind,l in enumerate(all_path):
    curr_path=path+'/'+l
    dom=ET.parse(curr_path)
    width=int(dom.findall('size/width')[0].text)
    height=int(dom.findall('size/height')[0].text)
    i=i+1
   # print(width,height)
    mask= np.zeros((height,width, 3), dtype = np.uint8)
    flag=False
    xmin=[]
    ymin=[]
    xmax=[]
    ymax=[]
   # print(l)
    number_of_column=len(dom.findall('object/bndbox/xmin'))
    print(number_of_column)
    if number_of_column>0:
        for g in range(number_of_column):
            xmin.append(int(dom.findall('object/bndbox/xmin')[g].text))
            ymin.append(int(dom.findall('object/bndbox/ymin')[g].text))
            xmax.append(int(dom.findall('object/bndbox/xmax')[g].text))
            ymax.append(int(dom.findall('object/bndbox/ymax')[g].text))
            flag=True
    if flag==True:
        for i in range(number_of_column):
            cv2.rectangle(mask, (xmin[i],ymin[i]), (xmax[i],ymax[i]),(255,255,255), -1)
    filename=l.split('.')[0]+'.jpg'
    cv2.imwrite(filename, mask)

#     if number_of_table>0:
#         for g in range(number_of_table):
#             ymin.append(int(dom.findall('object/bndbox/ymin')[g].text))
#             flag=True
#     if number_of_table>0:
#         for g in range(number_of_table):
#             xmax.append(int(dom.findall('object/bndbox/xmax')[g].text))
#             flag=True
#     if number_of_table>0:
#         for g in range(number_of_table):
#             ymax.append(int(dom.findall('object/bndbox/ymax')[g].text))
#             flag=True
#     if flag==True:
#         for i in range(number_of_table):
#             cv2.rectangle(mask, (xmin[i],ymin[i]), (xmax[i],ymax[i]),(255,255,255), -1)
#     filename=l.split('.')[0]+'.jpg'
#     cv2.imwrite(filename, mask)
```
